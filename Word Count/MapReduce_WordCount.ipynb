```python
def mapper(line):
    for word in line.strip().split():
        yield (word, 1)
# yields a tuple containing the current word and the value 1
```

```python
def reducer(word, values):
    return (word, sum(values))
# sum the values of each word
```

```python
with open("Football Analytics With Python & R Learning Data Science Through the Lens of Sports.txt", "r") as f:
    data = f.readlines()
    print(data)
# read my .txt data line by line
```

```python
import re
for i in range(len(data)):
    data[i] = re.sub(r"[^\w\s]+|\s+", " ", data[i].strip())

print(data)
# remove the extra \n and extra spaces form my data
```

```python
mapped_data = []
for line in data:
  mapped_data.extend(mapper(line))
# itrate over each line in the data and apply the mapper function on the line and pass the output(a word and 1) to the mapped_data list
```

```python
mapped_data[1]
```

```python
grouped_data = {}
for word, count in mapped_data:
  grouped_data.setdefault(word, []).append(count)
# grouped_data deictionary for each word and it's values,
# with setdefault checks if the word isn't in grouped_data then it will append the word and the value
# if the word in grouped_data it'll add the number to it's list
```

```python
reduced_data = []
for word, values in grouped_data.items():
  reduced_data.append(reducer(word, values))
#reduced_data is a list contain each word and the summtion of it's values
```

```python
for word, count in reduced_data:
  print(f"{word}: {count}")
```

```python
reduced_data = sorted(reduced_data,reverse=True)
```
